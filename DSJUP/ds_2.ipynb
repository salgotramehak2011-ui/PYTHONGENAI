```python
import pandas as pd
```

```python
ser = pd.Series([23,56,44,32,10,100,46,89,91])
ser
```

```python
ser > 50
```

```python
ser[ser > 50]
```

```python
ser[(ser > 30) & (ser < 70)]
```

```python
# between 
ser[ser.between(32,70,inclusive = 'both')] # 'both','neither','left','right', ## 30 and 70 both included in output
```

```python
ser[ser.between(32,70,inclusive = 'neither')] # 30 and 70 not included in output
```

```python
ser[ser.between(32,70,inclusive = 'right')] # 70 not included in output
```

```python
ser[ser == 46]
```

```python
# isin() function 
ser[ser.isin([46])]
```
