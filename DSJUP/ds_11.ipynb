```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic') # titanic, flights, tips,irirs
titanic
```

```python
titanic.loc[:,'class'].value_counts()
```

```python
sns.countplot(data=titanic, x='class')
```

```python
sns.countplot(data=titanic, x='class', hue='survived')
```

```python
sns.countplot(data=titanic, x='class', hue='alive', order=['Third','Second','First'])
```

```python
sns.countplot(data=titanic, x='class', hue='alive', order=['Third','Second','First'], hue_order=['yes','no'], width=0.5, fill=True,
             color='yellow',palette='plasma') # fill= true by default
```

```python
c=sns.countplot(data=titanic, x='class', hue='alive', order=['Third','Second','First'], hue_order=['yes','no'], width=0.5, fill=True,
             color='yellow',palette='plasma')
c.set_xlabel('Passenger class')
c.set_ylabel('Count of passenger')
c.set_title('count of passenger in classes')
c.set_xticklabels(['3rd','2nd','1st'])
```
