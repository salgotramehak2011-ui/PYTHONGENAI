### Bar chart:-

```python
fruits = ["Apples","Banana","Grapes","cherry"]
sales = [400,300,350,450]

plt.bar(x=fruits, height=sales, width=0.5)# x-axis--category, y-axis--numerical values
plt.show()
```

```python
import seaborn as sns 
# sns.barplot(x=fruits, y=sales)
sns.barplot(x=sales, y=fruits)
```

```python
plt.bar(x=fruits, height=sales, width=0.5, color=['red','yellow','green','purple'])
plt.show()
```

```python
plt.bar(fruits,sales, width=0.5, color=['red','yellow','green','purple'])
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.title('Sales vs Fruits')
plt.show()
```

```python
# x-axis--category, y-axis--numerical values
plt.barh(fruits,sales, height=0.6, color=['red','yellow','green','purple'])
plt.xlabel('Sales')  
plt.ylabel('Fruits') 
plt.title('Sales vs Fruits')
plt.show()
```

```python
import numpy as np
subjects = ['Maths','Hindi','English','Science']
boys = [20,30,40,50]
girls = [25,32,45,52]

x = np.arange(len(subjects))

plt.bar(x - 0.2, boys, width=0.4, label='Boys')
plt.bar(x + 0.2,girls, width=0.4, label='Girls')

plt.xticks(x,subjects)
plt.legend()
plt.show()
```

```python
import numpy as np
subjects = ['Maths','Hindi','English','Science']
boys = [20,30,40,50]
girls = [25,32,45,52]

x = np.arange(len(subjects))

plt.bar(x - 0.2, boys, width=0.4, label='Boys') # x - 0.2-->shift left
plt.bar(x + 0.2, girls, width=0.4, label='Girls') # x + 0.2-->shift right

plt.xticks(x,subjects) # xtricks--> is used to set custom labels on the X-axis.(x-->position,[0,1,2], labels)
plt.legend()
plt.show()
```

### Scatter Chart:-

```python
import numpy as np
x = np.array([12,45,7,32,89,54,23,67,14,91])
y = np.array([99,31,72,56,19,88,43,61,35,77])
plt.scatter(x, y)
plt.show()
```

```python
x1 = np.array([12,45,7,32,89,54,23,67,14,91])
y1 = np.array([99,31,72,56,19,88,43,61,35,77])

x2 = np.array([13,51,17,32,80,51,20,67,14,91])
y2 = np.array([99,31,21,60,9,81,43,17,35,77])

plt.scatter(x1, y1, label='Group1', color='red')
plt.scatter(x2, y2, label='Group2', color='green')
plt.legend()
plt.show()
```

### Pie Chart:-

```python
country = ['India','Pakistan','Bangladesh','Nepal','Spain']
values = [200,180,230,100,120]

plt.pie(x=values, labels=country)
plt.show()
```

```python
plt.pie(x=values, labels=country, autopct="%1.1F%%")
plt.title('Countries Data')
plt.grid(True)
plt.show()
```

```python
labels = ['hey','hello','hey everyone','hi','namaste']
values = [10,20,30,40,50]
explode = [0,0.1,0,0,0]
color=['red','yellow','orange','pink','purple']
plt.pie(values,labels=labels,colors=color,autopct="%1.1F%%",explode=explode,startangle=90)
plt.show()
```
